```python
import numpy as np
import matplotlib.pyplot as plt

# Évente mért adatok (10 év)
delta_t = 1
time = np.arange(0, 10, delta_t)  # 0-tól 9-ig, összesen 10 adatpont

# Népesség alakulása: 100 főről indul, 5%-os éves növekedéssel
population = 100 * (1.05) ** time

# Növekedési ütem: az évről évre bekövetkező abszolút növekedés
growth_rate = np.diff(population) / delta_t

# Következő évi (10. év) népesség előrejelzése:
# egyszerűen a legutolsó népességi értékhez hozzáadjuk a legutolsó növekedést
next_population = population[-1] + growth_rate[-1] * delta_t
# Alternatív megoldás: mivel a növekedési arány állandó, next_population = population[-1] * 1.05

print(f"A következő évi népszámláláson várható: {next_population:.2f} fő")

# Ábrázolás
plt.plot(time, population, 'bo-', label='Népesség')
plt.plot(time[:-1], growth_rate, 'r--', label='Növekedési ütem')
plt.xlabel('Idő (év)')
plt.ylabel('Érték')
plt.legend()
plt.title('Népesség növekedése és növekedési ütem')
plt.show()
```
