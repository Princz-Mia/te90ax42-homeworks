```python
import numpy as np
import matplotlib.pyplot as plt

m = 1.0  # tömeg (kg)
k = 50.0  # rugóállandó (N/m)
c = 0.5  # csillapítási tényező (Ns/m)
x0 = 0.1  # kezdeti kitérés (m)
v0 = 0.0  # kezdeti sebesség (m/s)
t_max = 100.0  # teljes idő (s)
delta_t = 0.1  # időlépés (s)

time = np.arange(0, t_max, delta_t)

x = x0
v = v0

positions = []
velocities = []
accelerations = []

for t in time:
    a = -(k * x + c * v) / m 
    positions.append(x)
    velocities.append(v)
    accelerations.append(a)
    
    v = v + a * delta_t
    x = x + v * delta_t

plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(time, positions, label="Kitérés (x)", color="blue")
plt.xlabel("Idő (s)")
plt.ylabel("Kitérés (m)")
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(time, velocities, label="Sebesség (v)", color="green")
plt.xlabel("Idő (s)")
plt.ylabel("Sebesség (m/s)")
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(time, accelerations, label="Gyorsulás (a)", color="red")
plt.xlabel("Idő (s)")
plt.ylabel("Gyorsulás (m/s²)")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
