```python
import numpy as np
import matplotlib.pyplot as plt

m = 1.0  # tmeg (kg)
k = 50.0  # ruglland (N/m)
c = 0.5  # csillaptsi tnyez (Ns/m)
x0 = 0.1  # kezdeti kitrs (m)
v0 = 0.0  # kezdeti sebessg (m/s)
t_max = 100.0  # teljes id (s)
delta_t = 0.1  # idlps (s)

time = np.arange(0, t_max, delta_t)

x = x0
v = v0

positions = []
velocities = []
accelerations = []

for t in time:
    a = -(k * x + c * v) / m 
    positions.append(x)
    velocities.append(v)
    accelerations.append(a)
    
    v = v + a * delta_t
    x = x + v * delta_t

plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(time, positions, label="Kitrs (x)", color="blue")
plt.xlabel("Id (s)")
plt.ylabel("Kitrs (m)")
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(time, velocities, label="Sebessg (v)", color="green")
plt.xlabel("Id (s)")
plt.ylabel("Sebessg (m/s)")
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(time, accelerations, label="Gyorsuls (a)", color="red")
plt.xlabel("Id (s)")
plt.ylabel("Gyorsuls (m/s)")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
